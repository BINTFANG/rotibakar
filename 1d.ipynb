```python
import matplotlib.pyplot as plt
import numpy as np

years = np.array([1981, 1983, 1985, 1987, 1989, 1991, 1993, 1995, 1997, 1999])
temperatures = np.array([14.1999, 14.2411, 14.0342, 14.2696, 14.197, 14.3055, 14.1853, 14.3577, 14.4187, 14.3438])

n = len(years)
Σx = np.sum(years)
Σy = np.sum(temperatures)
Σxy = np.sum(years * temperatures)
Σx2 = np.sum(years ** 2)

a = (n * Σxy - Σx * Σy) / (n * Σx2 - Σx ** 2)
b = (Σy - a * Σx) / n

regression_years = np.array([1981, 1999])
regression_temperatures = a * regression_years + b

linear_interpolation_years = np.arange(1982, 1998, 2)
linear_interpolation_temperatures = a * linear_interpolation_years + b

quadratic_interpolation_temperatures = (
    14.1999 + 0.0412 * 2 - 0.12405 * 2 * (2 + 1)
    + 0 * (2 ** 2)
)

cubic_interpolation_temperatures = (
    14.1999 + 0.0412 * 2 - 0.12405 * 2 * (2 + 1)
    + 0.03808 * (2 ** 2) * (2 + 1) * (2 + 1 + 2)
)

quadratic_interpolation_temperatures = np.full_like(
    linear_interpolation_years, quadratic_interpolation_temperatures
)

cubic_interpolation_temperatures = np.full_like(
    linear_interpolation_years, cubic_interpolation_temperatures
)

plt.figure(figsize=(10, 10))
plt.scatter(years, temperatures, color='red', label='Given Data')
plt.plot(regression_years, regression_temperatures, color='blue', label='Regression')
plt.plot(linear_interpolation_years, linear_interpolation_temperatures, color='green', label='Linear Interpolation')
plt.plot(linear_interpolation_years, quadratic_interpolation_temperatures, color='purple', label='Quadratic Interpolation')
plt.plot(linear_interpolation_years, cubic_interpolation_temperatures, color='orange', label='Cubic Interpolation')

plt.xlabel('Year')
plt.ylabel('Temperature (°C)')
plt.title('Relationship between Temperature and Year')
plt.legend()
plt.grid(True)
plt.show()
```
